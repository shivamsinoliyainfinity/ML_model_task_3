```python
data['Levy_missing']=np.where(data['Levy']=='-',1,0)
```

```python
data['Levy']=data['Levy'].replace('-', '0')
```

```python
data['Levy']=data['Levy'].astype(int)
```

```python
Levy_median = data.Levy.median()
```

```python
data['Levy'] = data['Levy'].replace(0, Levy_median)
```

```python
data['Engine volume']=data['Engine volume'].str.replace(' Turbo','')
data['Engine volume']=data['Engine volume'].astype(float)
```

```python
data['Mileage']=data['Mileage'].str.replace(' km', '').astype(int)
```

```python
data['Doors']=data['Doors'].str.replace('04-May','4-5').replace('02-Mar','2-3')
```

```python
num_data = data[['Price','Levy','Prod. year','Mileage','Cylinders','Airbags']]
```

```python
cat_data=data[['Manufacturer', 'Model', 'Category', 'Leather interior', 'Fuel type', 'Gear box type', 'Drive wheels', 'Doors', 'Wheel', 'Color']]
```

```python
for feature in cat_data.columns:
    cats_list=cat_data[feature].value_counts().sort_values(ascending=False).head(10).index
    cats_list=list(cats_list)
    for categories in cats_list:
        cat_data[categories]=np.where(cat_data[feature]==categories,1,0)
    cat_data.drop(columns=[feature], inplace=True)
cat_data.head()
```

```python
final_df = pd.concat([num_data, cat_data], axis=1)
final_df.head()
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
X = list(final_df.columns)
X.remove('Price')
target = ['Price']
X = final_df[X]
Y = final_df[target]
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.2, random_state=1)
```

```python
model = RandomForestRegressor()
model.fit(train_X,train_Y)
```

```python
prediction = model.predict(test_X)
```

```python
model.score(test_X,test_Y)
```

```python
from sklearn import metrics
metrics.r2_score(test_Y, prediction)
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(test_Y,prediction)
mae
```
